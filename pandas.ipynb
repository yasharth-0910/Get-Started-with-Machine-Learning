```python
# Create a dictionary with some data
dict = {
    'A': [1, 2, np.nan],
    'B': [5, np.nan, np.nan],
    'C': [1, 2, 3]
}

# Create a DataFrame from a dictionary
df = pd.DataFrame(dict)

print(df)
```

Pandas has 2 Data Structures

1) Series - One-dimensional labeled array that can hold data of any type (integer, float, string, objects, etc.)
2) DataFrame - Two-dimensional labeled data structure with columns that can hold different types of data




```python
new = pd.Series(np.random.rand(5), index = ["a", "b", "c", "d", "e"]) # Creating a new series, with random values and custom index values

print(new) 
```

```python
newdf = pd.DataFrame(np.random.rand(300,5), index = np.arange(300), columns = ["A", "B", "C", "D", "E"]) # Creating a new DataFrame, with random values and custom index values
#300 rows, 5 columns with custom index values and column names
print(newdf)

newdf.loc[0, "A"] = np.nan # Changing the value of a cell to NaN

newdf.loc[1, "A"] = np.nan # Changing the value of a cell to NaN
```

```python
newdf.to_csv("newdf.csv")

#newdf.to_excel("newdf.xlsx")

#newdf.to_json("newdf.json")

newfd_csv = pd.read_csv("newdf.csv")


#change name of columns

# Skip the first column (index) when renaming
newfd_csv = newfd_csv.set_index(newfd_csv.columns[0])
newfd_csv.columns = ["a", "b", "c", "d", "e"]


newfd_csv


```

```python
newfd_csv.to_numpy() # Convert DataFrame to NumPy array
```
